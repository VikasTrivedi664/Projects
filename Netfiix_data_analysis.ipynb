```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as px
from datetime import datetime
```

```python
df = pd.read_csv('Netflix.csv')
df.head()
```

```python
sns.set(rc={'figure.figsize':(10,5)})
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')
df.head()
```

 ----------------------------------------
 # 1.Volume of Stock Traded

```python
sns.lineplot(x = df.index, y = df['Volume'], label = 'Volume')
plt.title('Stock Verses Time')
```

----------------------
# 2.Netflix Stock Price - High , Open , Close

```python
df.plot(y=['High','Close','Open'],title= 'Netflix Stock Price High,Open and Close')
```

----------------------------------
# 3.Netflix Stock Price - Day, Month, Year wise

```python
fig,(ax1,ax2,ax3) = plt.subplots(3,figsize = (12,12))
df.groupby(df.index.day).mean().plot(y = 'Volume',ax = ax1,xlabel = 'Day',title = 'Stock Analysis Day-by-day')
df.groupby(df.index.month).mean().plot(y = 'Volume',ax = ax2,xlabel = 'Month',title = 'Stock Analysis Month-by-Month')
df.groupby(df.index.year).mean().plot(y = 'Volume',ax = ax3,xlabel = 'Year',title = 'Stock Analysis Year-by-Year')
```
